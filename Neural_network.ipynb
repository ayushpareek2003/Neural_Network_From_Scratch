```python
#sigmoid for activation
def sig(y):
    return (1 / (1 + np.exp(-y)))
    
```

```python
#log loss function or binarycrossentropy
def logl(a,b):
    eps = 1e-15
    y = [max(min(i, 1 - eps), eps) for i in b]
    y = np.array(y)
    
    return -np.mean(a * np.log(b) + (1 -a) * np.log(1 - y))
    
```

```python
def mode(a,b,c,d,e,f,output,epochs):
        w1=1
        w2=1
        w3=1
        w4=1
        w5=1
        w6=1
        bias=1
        rate=0.1
        
    
        n=len(a)
        for i in range(0,epochs):
            sum=w1*a + w2*b +w3*c +w4*d + w5*e+ w6*f +bias
            y_ped=sig(sum)
            loss=logl(output,y_ped)
            
            w1=w1-rate*((1/n)*(np.dot(np.transpose(a),(y_ped-output))))
            w2=w2-rate*((1/n)*(np.dot(np.transpose(b),(y_ped-output))))
            w3=w3-rate*((1/n)*(np.dot(np.transpose(c),(y_ped-output))))
            w4=w4-rate*((1/n)*(np.dot(np.transpose(d),(y_ped-output))))
            w5=w5-rate*((1/n)*(np.dot(np.transpose(e),(y_ped-output))))
            w6=w6-rate*((1/n)*(np.dot(np.transpose(f),(y_ped-output))))
            bias=bias-rate*(np.mean(y_ped-output))

            print('Epoch',i+1,'LOSS',loss)
        return w1,w2,w3,w4,w5,w6,bias 
```

```python
a,b,c,d,e,f,w=mode(data['Calories Burn'],data['Age'],data['Duration'],data['Heart Rate'],data['BMI'],data['Tar'],Target,100)
```
